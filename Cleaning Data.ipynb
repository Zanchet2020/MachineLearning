```python
%matplotlib inline
import numpy as np

incomes = np.random.normal(27000, 15000, 100000)
incomes = np.append(incomes, np.random.uniform(0, 27000 * 2, 10))

import matplotlib.pyplot as plt

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.mean()
```

```python
def reject_outliers(data):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - 2 * s < e < u + 2 * s)]
    return filtered
```

```python
filtered = reject_outliers(incomes)

plt.hist(filtered,50)
plt.show()
```

```python
np.mean(filtered)
```
